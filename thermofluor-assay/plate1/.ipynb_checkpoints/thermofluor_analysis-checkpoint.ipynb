```python
import pandas as pd
import seaborn as sns
import matplotlib 
import matplotlib.pyplot as plt 
import re
from matplotlib.offsetbox import AnchoredText
sns.set_style("white")
%matplotlib inline 
```

```python
df = pd.read_csv("plate1/high_expression/plate1_high_expression_fluorescence_vs_temperature.csv")
```

```python
df_tm = pd.read_csv('plate1/high_expression/plate1_high_expression_manual_Tm_detection.csv', header=1)
```

```python
temps = list(df['X'])
```

```python
col_to_keep = [x for x in df.columns if not re.match('X', x)]
df_data = df[col_to_keep]
```

```python
df4= pd.read_csv("plate4/high_expression/plate4_high_expression_fluorescence_vs_temperature.csv")
col_to_keep = [x for x in df4.columns if not re.match('X', x)]
df_data4 = df4[col_to_keep]
df_tm4 = pd.read_csv('plate4/high_expression/plate4_high_expression_manual_Tm_detection.csv', header=1)
temps4 = list(df4['X'])
```

```python
fig = plt.figure(figsize=(20,20), dpi=300)
for i,col in enumerate(col_to_keep):
    index = i +1
    ax = fig.add_subplot(5,3,index)
    plt.plot(temps, df_data[col], color='red')
    #plt.plot(temps4, df_data4[col], color='orange')
    plt.axvline(x=list(df_tm['Tm1'])[i], color='black', linestyle='--')
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    if i == 13: 
        plt.xlabel('Temperature ($^\circ$C)', fontsize=50)
    if i == 6: 
        plt.ylabel('Fluorescence (AU)', fontsize=50)
    #anchored_text = AnchoredText("$T_m^1$ : %.0f$^\circ$C \n $T_m^2$ : %.0f$^\circ$C " % (list(df_tm['Tm1'])[i], list(df_tm4['Tm1'])[i]) , loc=1, prop=dict(size=15))
    anchored_text = AnchoredText("$T_m$ : %.0f$^\circ$C" % (list(df_tm['Tm1'])[i]) , loc=1, prop=dict(size=15))
    ax.add_artist(anchored_text)
fig.savefig('plate1_tm.pdf', dpi=300)
```

```python
fig = plt.figure(figsize=(20,20), dpi=300)
for i,col in enumerate(col_to_keep):
    index = i +1
    ax = fig.add_subplot(5,3,index)
    plt.plot(temps, df_data[col], color='blue')
    plt.plot(temps4, df_data4[col], color='green')
    plt.axvline(x=list(df_tm['Tm1'])[i], color='black', linestyle='--')
    plt.axvline(x=list(df_tm4['Tm1'])[i], color='black', linestyle=':')
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    if i == 13: 
        plt.xlabel('Temperature ($^\circ$C)', fontsize=50)
    if i == 6: 
        plt.ylabel('Fluorescence (AU)', fontsize=50)
    anchored_text = AnchoredText("$T_m^{b}$ : %.0f$^\circ$C \n $T_m^{g}$ : %.0f$^\circ$C " % (list(df_tm['Tm1'])[i], list(df_tm4['Tm1'])[i]) , loc=1, prop=dict(size=15))
    #anchored_text = AnchoredText("$T_m^1$ : %.0f$^\circ$C" % (list(df_tm['Tm1'])[i]) , loc=1, prop=dict(size=15))
    ax.add_artist(anchored_text)
fig.savefig('bothplates_tm.pdf', dpi=300)
```
