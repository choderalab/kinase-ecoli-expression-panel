```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_excel('abl1-constructs-with-boundaries.xlsx')
```

```python
df
```

```python
new_df = df.drop_duplicates()
```

```python
df_pivot = new_df.pivot('N Terminus', 'C Terminus', 'yield')
```

```python
fig = plt.figure(figsize=[6,4])
cmap = sns.cubehelix_palette(as_cmap=True, rot=-.3, light=1)
ax = sns.heatmap(df_pivot, robust=True, cmap=cmap, linewidths=.5, cbar_kws={'label': 'Yield ($\mu$g/mL)'})
ax.invert_yaxis()
plt.yticks(fontsize=12, rotation='horizontal')
plt.xticks(fontsize=12)
plt.ylabel('N Terminus', fontsize=15)
plt.xlabel('C Terminus', fontsize=15)
plt.savefig('heatmap-construct.pdf', dpi=500)
```
